```python
import pandas as pd

pd.set_option('display.max_columns', None)
```

```python
df_loaded = pd.read_pickle(r'../CMU pkl file/films.pkl')

df_raw = pd.DataFrame([(k, v['genres'], v['summary'],) for k, v in df_loaded.items()],
             columns=['ID','GENRES','SUMMARY']
            )

print(df_raw.shape)

df_raw = df_raw[df_raw.SUMMARY != '']

print(df_raw.shape)
```

First level replacement based on manual mapping sheet

```python
# Most movies have one genre, but there are quite a few with up to 5. Clearly this is a multi-label probelm

df_raw['NUM_GENRES'] = df_raw.apply(lambda row: len(row.GENRES),axis=1)

df_raw['NUM_GENRES'].value_counts(normalize=True)
```

```python
df_raw[df_raw['NUM_GENRES']==1]
```

```python
# there are 363 different genres. Let's try to reduce some of these.

extracted_genres = []

for i in range(0,len(df_raw)):
    for j in range(0,df_raw.iloc[i]['NUM_GENRES']):
        extracted_genres.append(df_raw.iloc[i]['GENRES'][j])
        
all_genres = sorted(list(set(extracted_genres)))

len(all_genres)        
```

```python
# here we manually create a mapping sheet
#pd.DataFrame(all_genres).to_csv(r'./All_Genres.csv')
df_mapping = pd.read_csv(r'./All_Genres_Mapped.csv') # load mapping sheet
```

```python
#df_mapping.AMENDED.value_counts()[df_mapping.AMENDED.value_counts()==1].to_frame().reset_index()['index'].tolist()
```

```python
mapping_dict = dict(zip(df_mapping.ORIGINAL, df_mapping.AMENDED))
```

```python
mapping_dict
```

```python
new_col = []
for i in range(0,len(df_raw)):
    new_col_val = [] 
    for j in range(0,df_raw.iloc[i]['NUM_GENRES']):
        new_col_val.append(mapping_dict[df_raw.iloc[i]['GENRES'][j]])
    new_col.append(list(set(new_col_val)))
```

```python
df_raw['CLEANED_GENRES'] = new_col
```

```python
df_raw
```

```python
# Most movies have one genre, but there are quite a few with up to 5. Clearly this is a multi-label probelm

df_raw['NUM_CLEANED_GENRES'] = df_raw.apply(lambda row: len(row.CLEANED_GENRES),axis=1)

df_raw['NUM_CLEANED_GENRES'].value_counts(normalize=True)
```

```python
# there are now 250 different genres. Let's try to reduce some of these.

extracted_genres = []

for i in range(0,len(df_raw)):
    for j in range(0,df_raw.iloc[i]['NUM_CLEANED_GENRES']):
        extracted_genres.append(df_raw.iloc[i]['CLEANED_GENRES'][j])
        
all_genres = sorted(list(set(extracted_genres)))

len(all_genres)
```

```python
import collections
genre_movie_counts = dict(collections.Counter(extracted_genres))
```

```python
sorted_dict = sorted(genre_movie_counts.items(), key=lambda x: x[1], reverse=True)
```
